```python
import numpy as np 
import matplotlib.pyplot as plt
a=np.linspace(0,10,20)
b=np.linspace(10,0,20)
plt.figure(figsize = (4,4))
ax1 = plt.subplot(121)  # 1= row , 2= cloumn , 1= no. of position 
ax2 = plt.subplot(122)
ax1.plot(a,linewidth = 1.0,color = 'g' , alpha = 1.0)
ax2.plot(b,linewidth = 1.0,color = 'g' , alpha = 1.0)
plt.show()   #plot is to write to draw a line
```
